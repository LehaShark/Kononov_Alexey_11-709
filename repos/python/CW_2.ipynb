```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
series = np.empty((0,2))

with open("Data\sunspot.txt", "r") as file:
    data = file.readlines()
    for i, line in enumerate(data):
        if i>2 and i<3080:
            words = line.split(" ")
            if len(words) != 1:
                #print(float(words[-1][:-1]))
                series = np.append(series, [[i-2, float(words[-1][:-1])]], axis=0)
#print(series)

X = series[:,0]
Y = series[:,1]
```

```python
plt.figure(figsize = (20,10))
plt.plot(X,Y)
plt.scatter(X,Y)
plt.xlim(100,1000)

plt.show()
```

```python
def moving_average(series, windowsize):
    results = [series[0]]
    for n in range(1, len(series)-windowsize):
        results.append(sum(series[n+i] for i in range(windowsize))/float(windowsize))
    return results

Y_ma = moving_average(Y, 100) 

plt.figure(figsize=(20,10))
plt.plot(X,Y)
plt.scatter(X,Y)
plt.plot(X[:len(Y_ma)],Y_ma, c="red", linewidth = 5)
plt.xlim(100,1000)
plt.show()
```
