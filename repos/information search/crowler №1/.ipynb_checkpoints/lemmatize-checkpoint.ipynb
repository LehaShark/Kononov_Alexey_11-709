```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()

def lemmatize(text):
    words = text.split()
    res = list()
    for word in words:
        p = morph.parse(word)[0]
        res.append(p.normal_form)

    return res

num = 0

while True:
    num = num + 1
    
    page = open(f"page{num}.txt", 'r', encoding='utf-8')
#f = open(f"page1.txt", 'r', encoding='utf-8')
    text = page.read()

    print(lemmatize(text))
```
